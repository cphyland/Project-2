```python
import pandas as pd
from pandas.io import sql
from sqlalchemy import create_engine
```

```python
import sqlite3
```

```python
sharks_df = pd.read_csv('data/sharks/shark_sightings_raw.csv')
sharks_df.head()
```

```python
# Create an engine that can talk to the database
engine = create_engine(f"sqlite:///{sharks_df}")
conn = engine.connect()
```

```python
sharks_df.to_sql('sharks_df', conn, if_exists='append', index=False)
```

```python
sharks_df = pd.read_sql("SELECT * FROM sharks_df", conn)
```
